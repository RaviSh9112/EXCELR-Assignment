```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import AgglomerativeClustering
from pandas.core.common import random_state
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(r"C:\Users\ravindra sharma\Desktop\excelrassignment\DataSet\crime_data.csv")
```

# Hierarchical Clustering


```python
df.head()
```

```python
#Normalizing the function
scaling=MinMaxScaler()
df_norm=scaling.fit_transform(df.iloc[:,1:])
```

```python
df_norm
```

```python
#Create Dendrogram
from scipy.cluster.hierarchy import linkage
Z = linkage(df_norm, method='average', metric= 'euclidean')
plt.figure(figsize=(15,5))
plt.title("Dendrogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(Z, leaf_rotation=0, leaf_font_size=8)
plt.show()
```

```python
#Create Cluster
hc=AgglomerativeClustering(n_clusters=4,affinity="euclidean",linkage='complete')
hc
```

```python
#Build cluster algorithm
y_huc=hc.fit_predict(df_norm)
cluster=pd.DataFrame(y_huc,columns=['cluster'])
```

```python
df['Cluster']=hc.labels_
```

```python
df.head(15)
```

```python
df.iloc[:,1:].groupby(df.Cluster).mean()
```

# DBSCAN
